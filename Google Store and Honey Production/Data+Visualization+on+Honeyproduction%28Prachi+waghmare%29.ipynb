## 3. Display the percentage distribution of the data in each year using the pie chart.

```python
# calculate the percentage distribution of the data in each year
year_perdist = df.groupby('year').size()/len(df)*100

# plotting the pie chart
plt.figure(figsize=(8,6))
plt.pie(year_perdist, labels=year_perdist.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Data in Each Year')
plt.show()
```

## 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings. 

```python
plt.figure(figsize=(8,6))
sns.distplot(df['priceperlb'])
plt.title("distribution of the variable (price per lb) using displot")
plt.show()
```

#### Inference
The distplot is utilized to illustrate the distribution of data, allowing for a clear observation of the price per pound data points. The price per pound represents the average price of honey per pound concerning sales. It can be inferred that as the price per pound increases, there is a corresponding decrease in sales.

## 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings

```python
plt.figure(figsize=(8,6))
plt.scatter(df['numcol'], df['prodvalue'],color = "tan")
plt.xlabel('Number of Colonies')
plt.ylabel('Production Value')
plt.title('Relationship between Number of Colonies and Production Value using a Scatter Plot.')
plt.show()
```

#### Inference
Based on the presented scatter plot, it can be observed that the variables numcol and prodval exhibit a positive correlation, suggesting that as the number of colonies increases, the production value of honey also tends to increase. Therefore, it can be inferred that an increase in the number of colonies is likely to result in an increased production of honey.






## 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings. 

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='year', y='prodvalue', data=df)
plt.title('Relationship between Year and Prodvalue')
plt.show()
```

#### Inference
The box plot is employed to provide a concise summary of the data. In this instance, the box plot was utilized to summarize the 'prodvalue' (product value) per year. An evident trend shows an increase in sales as the years progress, accompanied by a corresponding rise in the median sales. Notably, the range of the featured product value has expanded consistently every year. Specifically, the values have fluctuated within the range of 0.1 to 0.5 from the initial year in 1998 to the concluding year in 2012.

## 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
sns.pairplot(df, vars=['numcol', 'yieldpercol', 'totalprod', 'prodvalue', 'year'], hue='year')
```

##### Inference
The pairplot analysis visually illustrates both the univariate distributions of single variables and the bivariate relationships between two variables. It is a powerful tool for identifying correlations among variables. In this pairplot, the diagonal represents histograms displaying the distribution of individual variables, while the lower triangle exhibits scatter plots, and the upper triangle mirrors the lower.

The observations derived from the presented pairplot suggest direct relationships between certain pairs of variables: specifically, between numcol and totalprod, yieldpercol and prodvalue, and totalprod and prodvalue. Furthermore, it's evident that variables like numcol, totalprod, and prodvalue demonstrate a positively skewed nature in their distributions, while yieldpercol follows a relatively normal distribution pattern.
